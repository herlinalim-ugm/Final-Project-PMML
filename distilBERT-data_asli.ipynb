```python
import warnings
warnings.filterwarnings("ignore")

import os
import random
import numpy as np
import pandas as pd
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

from transformers import AutoTokenizer, AutoModel
from sklearn.model_selection import KFold
from sklearn.metrics import f1_score, precision_recall_fscore_support, precision_recall_curve

# ---------------------------
# CONFIG
# ---------------------------
TRAIN_CSV = "train.csv"
DEV_CSV   = "dev.csv"
TEXT_COL = "text"
LABEL_COLS = ["admiration","amusement","gratitude","love","pride","relief","remorse"]

MODEL_NAME = "distilbert-base-uncased"
MAX_LEN = 128
BATCH_SIZE = 64
EPOCHS = 20
LR = 5e-5
ACCUM_STEPS = 2
OUTPUT_DIR = "bert_kfold_out/"
os.makedirs(OUTPUT_DIR, exist_ok=True)
SEED = 42
PATIENCE = 10
N_FOLDS = 5

# ---------------------------
# Utils
# ---------------------------
def set_seed(seed=SEED):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

def multilabel_accuracy(y_true, y_pred):
    return (y_true == y_pred).all(axis=1).mean()

set_seed()

# ---------------------------
# Device
# ---------------------------
device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
device_type = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

# ---------------------------
# Load data
# ---------------------------
train_val_df = pd.read_csv(TRAIN_CSV)
test_df      = pd.read_csv(DEV_CSV)

for c in LABEL_COLS:
    train_val_df[c] = train_val_df[c].fillna(0).astype(int).clip(0,1)
    test_df[c]      = test_df[c].fillna(0).astype(int).clip(0,1)

# ---------------------------
# Tokenizer
# ---------------------------
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

def tokenize_texts(texts):
    enc = tokenizer(
        texts,
        padding="max_length",
        truncation=True,
        max_length=MAX_LEN,
        return_tensors="pt"
    )
    return enc["input_ids"], enc["attention_mask"]

# ---------------------------
# Model
# ---------------------------
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = AutoModel.from_pretrained(MODEL_NAME)
        try:
            self.backbone.gradient_checkpointing_enable()
        except Exception:
            pass
        hidden = self.backbone.config.hidden_size
        self.dropout = nn.Dropout(0.2)
        self.fc = nn.Linear(hidden, len(LABEL_COLS))

    def forward(self, ids, mask):
        out = self.backbone(ids, mask, return_dict=True)
        x = out.last_hidden_state[:,0,:]
        x = self.dropout(x)
        return self.fc(x)

# ---------------------------
# K-Fold Training
# ---------------------------
kf = KFold(n_splits=N_FOLDS, shuffle=True, random_state=SEED)
fold_results = []

for fold, (train_idx, val_idx) in enumerate(kf.split(train_val_df)):
    print(f"\n===== Fold {fold+1}/{N_FOLDS} =====")

    history = {
        "epoch": [],
        "train_loss": [],
        "val_loss": [],
        "train_acc": [],
        "val_acc": [],
        "val_f1": []
    }

    fold_train_df = train_val_df.iloc[train_idx].reset_index(drop=True)
    fold_val_df   = train_val_df.iloc[val_idx].reset_index(drop=True)

    train_ids, train_mask = tokenize_texts(fold_train_df[TEXT_COL].astype(str).tolist())
    val_ids, val_mask     = tokenize_texts(fold_val_df[TEXT_COL].astype(str).tolist())

    train_labels = torch.tensor(fold_train_df[LABEL_COLS].values, dtype=torch.float)
    val_labels   = torch.tensor(fold_val_df[LABEL_COLS].values, dtype=torch.float)

    train_dataset = TensorDataset(train_ids, train_mask, train_labels)
    val_dataset   = TensorDataset(val_ids, val_mask, val_labels)

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader   = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

    model = Model().to(device)

    pos_counts = train_labels.sum(dim=0)
    total_counts = train_labels.size(0)
    pos_weight = ((total_counts - pos_counts) / (pos_counts + 2e-5)).to(device)
    criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)

    optimizer = torch.optim.AdamW(model.parameters(), lr=LR)
    scaler = torch.amp.GradScaler()

    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode="max", factor=0.5, patience=1, min_lr=1e-7
    )

    best_f1 = 0.0
    best_thresholds = [0.5] * len(LABEL_COLS)
    patience_counter = 0

    for epoch in range(EPOCHS):
        model.train()
        optimizer.zero_grad()

        total_train_loss = 0
        train_preds, train_gts = [], []

        for i, batch in enumerate(tqdm(train_loader)):
            ids, mask, labels = [b.to(device) for b in batch]

            with torch.amp.autocast(device_type=device_type):
                logits = model(ids, mask)
                loss = criterion(logits, labels) / ACCUM_STEPS

            scaler.scale(loss).backward()

            if (i+1) % ACCUM_STEPS == 0:
                scaler.step(optimizer)
                scaler.update()
                optimizer.zero_grad()

            total_train_loss += loss.item() * ACCUM_STEPS
            preds = (torch.sigmoid(logits) > 0.5).int()
            train_preds.append(preds.cpu())
            train_gts.append(labels.cpu())

        train_loss = total_train_loss / len(train_loader)
        train_acc = multilabel_accuracy(
            torch.cat(train_gts).numpy(),
            torch.cat(train_preds).numpy()
        )

        # ---------- Validation ----------
        model.eval()
        total_val_loss = 0
        probs_all, labels_all = [], []

        with torch.no_grad():
            for batch in val_loader:
                ids, mask, labels = [b.to(device) for b in batch]
                logits = model(ids, mask)
                loss = criterion(logits, labels)

                total_val_loss += loss.item()
                probs_all.append(torch.sigmoid(logits).cpu())
                labels_all.append(labels.cpu())

        val_loss = total_val_loss / len(val_loader)
        probs_all = torch.cat(probs_all).numpy()
        labels_all = torch.cat(labels_all).numpy()

        thresholds = []
        for i in range(len(LABEL_COLS)):
            p, r, t = precision_recall_curve(labels_all[:, i], probs_all[:, i])
            f1s = 2*p*r / (p+r+1e-8)
            idx = np.argmax(f1s)
            thresholds.append(t[idx] if idx < len(t) else 0.5)

        preds_val = (probs_all > np.array(thresholds)).astype(int)
        val_f1 = f1_score(labels_all, preds_val, average="micro")
        val_acc = multilabel_accuracy(labels_all, preds_val)

        history["epoch"].append(epoch+1)
        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        history["train_acc"].append(train_acc)
        history["val_acc"].append(val_acc)
        history["val_f1"].append(val_f1)

        print(f"Epoch {epoch+1} | TrainLoss {train_loss:.4f} "
              f"| ValLoss {val_loss:.4f} | ValF1 {val_f1:.4f}")

        scheduler.step(val_f1)

        if val_f1 > best_f1:
            best_f1 = val_f1
            best_thresholds = thresholds
            torch.save({
                "model_state": model.state_dict(),
                "thresholds": best_thresholds
            }, os.path.join(OUTPUT_DIR, f"best_fold{fold}.pt"))
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= PATIENCE:
                break

    # ---------- Export CSV ----------
    history_df = pd.DataFrame(history)
    history_df.to_csv(
        os.path.join(OUTPUT_DIR, f"history_fold{fold}.csv"),
        index=False
    )

    # ---------- Plot ----------
    plt.figure(figsize=(14,5))
    plt.subplot(1,2,1)
    plt.plot(history["train_loss"], label="Train")
    plt.plot(history["val_loss"], label="Val")
    plt.title(f"Fold {fold+1} Loss")
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(history["train_acc"], label="Train")
    plt.plot(history["val_acc"], label="Val")
    plt.title(f"Fold {fold+1} Accuracy")
    plt.legend()
    plt.show()

    fold_results.append({
        "fold": fold,
        "model_path": os.path.join(OUTPUT_DIR, f"best_fold{fold}.pt"),
        "thresholds": best_thresholds
    })

# ===========================================================
# TEST EVALUATION (ENSEMBLE)  TETAP
# ===========================================================
test_ids, test_mask = tokenize_texts(test_df[TEXT_COL].astype(str).tolist())
test_labels = torch.tensor(test_df[LABEL_COLS].values, dtype=torch.float)
test_loader = DataLoader(
    TensorDataset(test_ids, test_mask, test_labels),
    batch_size=BATCH_SIZE,
    shuffle=False
)

probs_all_folds = []

for f in fold_results:
    model = Model().to(device)
    ckpt = torch.load(f["model_path"])
    model.load_state_dict(ckpt["model_state"])
    model.eval()

    probs = []
    with torch.no_grad():
        for ids, mask, _ in test_loader:
            logits = model(ids.to(device), mask.to(device))
            probs.append(torch.sigmoid(logits).cpu())
    probs_all_folds.append(torch.cat(probs).numpy())

probs_all = np.mean(probs_all_folds, axis=0)
avg_thresholds = np.mean([f["thresholds"] for f in fold_results], axis=0)
preds = (probs_all > avg_thresholds).astype(int)

labels_all = test_labels.numpy()
micro_f1 = f1_score(labels_all, preds, average="micro")
print("\n=== TEST MICRO F1 (EnSEMBLE) ===")
print(f"{micro_f1:.4f}")

precision, recall, f1, _ = precision_recall_fscore_support(labels_all, preds, zero_division=0)
print("\n=== PER LABEL METRICS ===")
for i,label in enumerate(LABEL_COLS):
    print(f"{label:10s}  P: {precision[i]:.3f} | R: {recall[i]:.3f} | F1: {f1[i]:.3f}")


labels_all = test_labels.numpy()
micro_f1 = f1_score(labels_all, preds, average="micro")

# ===========================================================
# CONFUSION MATRIX & HEATMAP PER LABEL
# ===========================================================
num_labels = len(LABEL_COLS)

fig, axes = plt.subplots(2, 4, figsize=(18, 9))
axes = axes.flatten()

for i, label in enumerate(LABEL_COLS):
    cm = confusion_matrix(labels_all[:, i], preds[:, i])

    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        cbar=False,
        ax=axes[i]
    )

    axes[i].set_title(label.capitalize(), fontsize=12)
    axes[i].set_xlabel("Predicted")
    axes[i].set_ylabel("Actual")

    axes[i].set_xticks([0.5, 1.5])
    axes[i].set_xticklabels(["0", "1"])
    axes[i].set_yticks([0.5, 1.5])
    axes[i].set_yticklabels(["0", "1"], rotation=0)

# Matikan subplot kosong (jika label < 8)
for j in range(num_labels, len(axes)):
    axes[j].axis("off")

plt.suptitle("Confusion Matrix per Emotion Label (Test Set)", fontsize=16)
plt.tight_layout()
plt.show()
```
